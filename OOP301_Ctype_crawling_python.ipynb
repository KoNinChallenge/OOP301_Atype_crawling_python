```python
### Header-import ###
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import time
import urllib.request
from tqdm import tqdm # image検索状況を確認のため
import os # imageを保存する場所（パス）を設定
import csv # csvファイルをリストに変換:　検索キーワードをより簡単に入力できる

# ここから「shift+enterキー」を押してください。
# 上記の方法でできない場合はメニューバーの「Kernel」をクリックし、「Restart & Run All」をクリックしてください。
```

```python
### クラス class ###

class AutoDownload():
    ### 初期化　Control: 変数・引数 -> 検索キーワード及び枚数選択等 ###
    def __init__(self, keyword, count_no, DEBUG_1 = False, DEBUG_2 = False, report = False): 
        
        self.keyword = keyword
        self.count_no = count_no
        self.DEBUG_1 = DEBUG_1
        self.DEBUG_2 = DEBUG_2
        self.debug_jump = DEBUG_2
        self.report = report
        
        # 追加したい設定があれば下記に記載してください。
        #***************** add attributes here ******************
        
        #********************************************************
        
#     def __repr__(self):
#         pass
    
    ### Main ###
    #実行
    def main(self):
        
        for keyword_i in self.keyword:
            self.keyword_i = keyword_i
            
            self.selenium_setup()
        
            # 変数設定
            driver = self.driver

            while True:
                scroll_stop = self.scroll_down()
                if scroll_stop:
                    break
                            
            self.img_folder()
            
            ## Pictureクリック及び保存
            self.img_save()

            #　最終の結果報告
            if keyword_i == keyword[-1]:
                print("終了: 最後のキーワードです。")

    ### Body ###
    ## Google Chromeで自動化
    def selenium_setup(self):
        
        #ドライブ設定:Google Chrome
        driver = webdriver.Chrome()
        driver.get("https://www.google.co.jp/imghp?hl=ja&ogbl")

        # Nameで検索
        elem = driver.find_element(By.NAME, "q")
        elem.send_keys(self.keyword_i)
        elem.send_keys(Keys.ENTER)

        self.driver = driver
           
    ## Scroll Downセット
    def scroll_down(self):
        
        # 変数設定
        driver = self.driver

        # 時間支援設定（秒）Time sleep setting(seconds)
        SCROLL_PAUSE = 1

        # 検索ページの高さをゲット　Get scroll height
        last_height = driver.execute_script("return document.body.scrollHeight")

        # スクロールダウンを自動化　Scroll down to bottom
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

        # タイミング調整　Wait to load page
        time.sleep(SCROLL_PAUSE)

        # 結果をもっと表示ボタンをクリック　Calculate new scroll height and compare with last scroll height
        new_height = driver.execute_script("return document.body.scrollHeight")
        if new_height == last_height:
            try:
                driver.find_element(By.CSS_SELECTOR, '.mye4qd').click()
            except:
                return True

        last_height = new_height
    
    # イメージの保存場所（パス）の指定
    def img_folder(self):
        
        # 変数設定
        keyword_i = self.keyword_i
        
        # 現在のパスを確認
        current_path = os.getcwd()
        current_path = current_path.replace("\\", "/")

        # フォルダを生成
        os.makedirs(name = f"{current_path}/{keyword_i}", exist_ok = True) 
        time.sleep(0.1)

        #　結果報告
        print(f"{current_path}/{keyword_i}に{keyword_i}のフォルダを生成しました。")

        img_path = f"{current_path}/{keyword_i}"

        self.img_path = img_path
    
    ## Debug : Debugは大事なのでcheckが必要
    def debug(self, e=False):
        
        # 変数設定
        driver = self.driver
        
        # selenium errorを確認
        if self.DEBUG_1:
            if e:
                print('==== エラー内容 ====')
                print('type:' + str(type(e)))
                print('e自身:' + str(e)) 

        #ブロック、著作権等の対応 デバッグ
        if DEBUG_2:
            current_url = driver.current_url
            debug_url = current_url.split("img#imgrc=")[-1]

            self.current_url = current_url
            self.debug_url = debug_url
        
    ## Pictureクリック及び保存
    def img_save(self):
        
        # 変数設定
        driver = self.driver
        count = 1

        # イメージをclassで設定
        images = driver.find_elements_by_css_selector('.wXeWr.islib.nfEiy')

        # イメージのダウンロードを自動化
        print(f"「{self.keyword_i}」開始")

        for image in tqdm(images):
            try:
                image.click()

                #ブロック、著作権等の対応 デバッグ
                self.debug(0)
                if DEBUG_2:
                    if self.debug_url in self.debug_jump:

                        print("+++JUMP!+++", "URL: "+ self.debug_url)
                        continue
                            
                time.sleep(2)

                imgURL = driver.find_element_by_xpath("/html/body/div[2]/c-wiz/div[3]/div[2]/div[3]/div[2]/div/div[2]/div[2]/div[2]/c-wiz/div/div/div/div[3]/div[1]/a/img[1]").get_attribute('src')

                img_path_file = f'{self.img_path}/{self.keyword_i}{str(count)}.jpg'

                # デバッグ処理　403 forbidden Debug
                opener=urllib.request.build_opener()
                opener.addheaders=[('User-Agent','Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/36.0.1941.0 Safari/537.36')]
                urllib.request.install_opener(opener)

                urllib.request.urlretrieve(imgURL, img_path_file)

                count += 1

                # 経過レポート half report
                if count == int(self.count_no/2):
                    print("***")
                    print(f"「{self.keyword_i}」イメージを合計{self.count_no}枚の半分位【{int(self.count_no/2)}枚】完了しました。")
                    print("***")

                # 指定枚数による停止及び終了　count stop
                if count > self.count_no:
                    print("*****************************")
                    print(f"おめでとうございます！ \n「{self.keyword_i}」イメージを【{self.count_no}枚】⇒ 保存（格納）しました。")
                    print("*****************************")
                    break

            except Exception as e:
                #　デバッグ報告
                self.debug(e)
                pass

        #　検索ブラウザのクローズ
        driver.close()
                
#     # 報告　report
#     def report_sns(self):
#         pass
```
