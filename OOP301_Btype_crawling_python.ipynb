```python
### Header-import ###
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import time
import urllib.request
from tqdm import tqdm # image検索状況を確認のため
import os # imageを保存する場所（パス）を設定

# ここから「shift+enterキー」を押してください。
# 上記の方法でできない場合はメニューバーの「Kernel」をクリックし、「Restart & Run All」をクリックしてください。
```

```python
### Control: 変数・引数 -> 検索キーワード及び枚数選択等 ###

## Debug 問題があった場合に設定
# debug1: selenium errorを確認
DEBUG_1 = False

# debug2: ブロック、著作権等の対応
DEBUG_2 = True
debug_jump = ["sXYsCDP-tN4bXM", "CTEglNCX2nMiQM","glQ6t9A1zWWksM"]#NG urlの"img#imgrc="以降をstrで記入

# B typeはinputを使わず、ここに直接入力する
keyword = ["玉ねぎ"] #ここにキーワードをstrで記入
count_no = 300 #イメージの枚数を入力する。
```

```python
### Body ###
## Google Chromeで自動化
def selenium_setup(keyword_i):
    #ドライブ設定:Google Chrome
    driver = webdriver.Chrome()
    driver.get("https://www.google.co.jp/imghp?hl=ja&ogbl")

    # Nameで検索
    elem = driver.find_element(By.NAME, "q")
    elem.send_keys(keyword_i)
    elem.send_keys(Keys.ENTER)
    
    return driver
```

```python
## Scroll Downセット
def scroll_down(driver):
    
    # 時間支援設定（秒）Time sleep setting(seconds)
    SCROLL_PAUSE = 1

    # 検索ページの高さをゲット　Get scroll height
    last_height = driver.execute_script("return document.body.scrollHeight")
    
    # スクロールダウンを自動化　Scroll down to bottom
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

    # タイミング調整　Wait to load page
    time.sleep(SCROLL_PAUSE)

    # 結果をもっと表示ボタンをクリック　Calculate new scroll height and compare with last scroll height
    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        try:
            driver.find_element(By.CSS_SELECTOR, '.mye4qd').click()
        except:
            return True

    last_height = new_height
```

```python
## Pictureクリック及び保存
def img_save(driver, img_path):
    # イメージをclassで設定
    images = driver.find_elements_by_css_selector('.wXeWr.islib.nfEiy')
    
    # イメージのダウンロードを自動化
    count = 1
    print(f"「{keyword_i}」開始")
    
    for image in tqdm(images):
        try:
            image.click()
            
            #ブロック、著作権等の対応 デバッグ
            if DEBUG_2:
                debug_url, _ = debug(driver, e=False)

                if debug_url in debug_jump:
                    print("+++JUMP!+++", "URL: "+debug_url)
                    continue
            
            time.sleep(2)

            imgURL = driver.find_element_by_xpath("/html/body/div[2]/c-wiz/div[3]/div[2]/div[3]/div[2]/div/div[2]/div[2]/div[2]/c-wiz/div/div/div/div[3]/div[1]/a/img[1]").get_attribute('src')
           
            img_path_file = f'{img_path}/{keyword_i}{str(count)}.jpg'

            # デバッグ処理　403 forbidden Debug
            opener=urllib.request.build_opener()
            opener.addheaders=[('User-Agent','Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/36.0.1941.0 Safari/537.36')]
            urllib.request.install_opener(opener)

            urllib.request.urlretrieve(imgURL, img_path_file)

            count += 1
            
            # 経過レポート half report
            if count == int(count_no/2):
                print("***")
                print(f"「{keyword_i}」イメージを合計{count_no}枚の半分位【{int(count_no/2)}枚】完了しました。")
                print("***")

            # 指定枚数による停止及び終了　count stop
            if count > count_no:
                print("*****************************")
                print(f"おめでとうございます！ \n「{keyword_i}」イメージを【{count_no}枚】⇒ 保存（格納）しました。")
                print("*****************************")
                break

        except Exception as e:
            #　デバッグ報告
            debug(driver, e)
            pass
    
    #　検索ブラウザのクローズ
    driver.close()
```

```python
# イメージの保存場所（パス）の指定
def img_folder():
    # 現在のパスを確認
    current_path = os.getcwd()
    current_path = current_path.replace("\\", "/")

    # フォルダを生成
    os.makedirs(name = f"{current_path}/{keyword_i}", exist_ok = True) 
    time.sleep(0.1)
    
    #　結果報告
    print(f"{current_path}/{keyword_i}に{keyword_i}のフォルダを生成しました。")
    
    img_path = f"{current_path}/{keyword_i}"
    
    return img_path
```

```python
## Debug : Debugは大事なのでcheckが必要
# Selenium Debug Check
def debug(driver, e=False):
    # selenium errorを確認
    if DEBUG_1:
        if e:
            print('==== エラー内容 ====')
            print('type:' + str(type(e)))
            print('e自身:' + str(e)) 
            
    #ブロック、著作権等の対応 デバッグ
    if DEBUG_2:
        current_url = driver.current_url
        debug_url = current_url.split("img#imgrc=")[-1]

        return debug_url, current_url
```

```python
### Main ###
#実行
for keyword_i in keyword:

    driver = selenium_setup(keyword_i)

    while True:
        scroll_stop = scroll_down(driver)
        if scroll_stop:
            break
            
    # イメージの保存場所の指定
    img_path = img_folder()
    
    ## Pictureクリック及び保存
    img_save(driver, img_path)
    
    #　最終の結果報告
    if keyword_i == keyword[-1]:
        print("終了: 最後のキーワードです。")
```
